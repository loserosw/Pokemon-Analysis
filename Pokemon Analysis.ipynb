```python
import pandas as pd
pokemon = pd.read_csv("Pokemon.csv")
pokemon
```

```python
import seaborn as sns
ax = sns.countplot(data=pokemon, x="Generation", hue="Generation", palette="bright")
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
for i in ax.containers:
    ax.bar_label(i,)
```

```python
total_pokemon = pokemon.nlargest(20, "Total")
total_pokemon
```

```python
ato = sns.barplot(data=total_pokemon, x="Total", y="Name", hue="Generation", palette="bright")
sns.move_legend(ato, "upper left", bbox_to_anchor=(1, 1))
for i in ato.containers:
    ato.bar_label(i,)
```
